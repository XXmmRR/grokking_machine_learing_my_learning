```python
input = 0.5
goal_pred = 0.8
weight = 0.1

for iteration in range(125):
    pred = input * weight
    error = (pred - goal_pred) ** 2
    delta = pred - goal_pred
    weight_delta = delta * input
    weight -= weight_delta
    
    print(f'Prediction {pred} Error{error} {weight}')
    
    
```
